```python
# baked_potato
import os
import pandas as pd
from PIL import Image
import io
# path to destination folder
folder_path = 'baked_potato'
# data we want to makw a dataframe from it
data = []
# go throught the list and give photo and it's name
for filename in os . listdir(folder_path):
    if filename . endswith(".jpg") or filename . endswith(".png"):  # Add more extensions if needed
        # Full path to the image
        img_path = os . path . join(folder_path, filename)
        # Open the image
        with open(img_path, 'rb') as img_file:
            img = img_file . read()
        # append each item into the dataframe
        data . append((filename, img))
# final
# make dataframe
df1 = pd . DataFrame(data, columns = ['Photo_Name', 'Photo'])
df1['Label'] = 'baked_potato'
df1 . to_csv('baked_potato.csv', index=False)
```
